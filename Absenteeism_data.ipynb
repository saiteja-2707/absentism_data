# Droping function

```python
df = df.drop(columns=['ID']) #drop function is used to drop unwanted column in we are dropinng the "ID" column
df                             #and storing it to df
```

```python
df = df.drop(columns=['Age'])   #drop function is used to drop unwanted column in we are dropinng the "AGE" column
df                              #and storing it to df
```

# To know the maximum and minimum value by using max()&min()

```python
print("largest value in Reason for Absence is:",df['Reason for Absence'].idxmax()) #gives the output of max value
```

```python
print("smallest value in Reason for Absence is:",df['Reason for Absence'].idxmin()) #gives the output of min value
```

# Data visualisation

```python
import matplotlib.pyplot as plt
x=df['Reason for Absence']           #x-axis as Reason for Absence
y=df['Absenteeism Time in Hours']    #y-axis as Absenteeism Time in Hours
plt.plot(grid=True)
plt.plot(x,y)                       #plt.plot() is used for line graph
plt.plot(x,y)
plt.legend()
plt.show()                           #to show the graph
```

```python
#bar graph for two columns in dataset

plt.title("Total attendence")               #plt.title is used title of the graph 
x=df['Reason for Absence']                  #x-axis
y=df['Absenteeism Time in Hours']           #y-axis
plt.bar(x,y)                                #plt.bar() is for creating bar graph
plt.xlabel("Reason for Absence")            #plt.xlabel for labeling the graph on x-axis
plt.ylabel("Absenteeism Time in Hours")     #plt.ylabel for labeling the graph on y-axis
plt.show()                                   #for showing the graph plt.show()
```
